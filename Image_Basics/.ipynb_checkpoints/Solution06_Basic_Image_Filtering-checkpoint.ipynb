# __Basic Image Filtering__
 
## Filtering in the spatial domain

The goal of theses exercises is to become familiar with the basic spatial image filtering algorithms, including linear and nonlinear filters. At first, we import the usual stuff and some filtering specific functions:

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
#import scipy.ndimage.filters as filters
from scipy.ndimage import filters
import time

np.set_printoptions(precision=4)
%matplotlib inline

plt.rcParams['font.size'] = 12
```

## A. Read in the Shepp-Logan phantom 512x512 and add some white noise

```python
SL = mpimg.imread('data/phantom.png');
sx = SL.shape[0]; sy = SL.shape[1];# Size of the image
# And add some normal noise, with standard deviation of 0.25 and 0 mean (np.random.randn)
r = 0.25 * np.random.randn(sx,sy);
SLn = SL + r;
plt.figure(1); plt.subplot(121); plt.imshow(SL,cmap=plt.cm.gray); plt.axis('off'); plt.colorbar(); plt.title('Original');
plt.figure(1);plt.subplot(122); plt.imshow(SLn,cmap=plt.cm.gray); plt.axis('off'); plt.colorbar();plt.title('Corrupted');
```

# B. Smoothing

B.1. Create the average filter M_a of size 2*N+1, with N ranging from 1 to 5
and define the bordering conditions set to 'mirror';

Apply the filter to the noisy version of the image using convn
and compare the output SLn_f to the original image

```python
for N in range(1,6):
    M_a = np.ones((N,N)); # Average Mask
    SLn_a = filters.convolve(SLn,M_a,mode='reflect');
    plt.figure(1); plt.subplot(121); plt.imshow(SLn_a,cmap=plt.cm.gray); plt.axis('off');plt.title('Average Filter -N = %d' %N);
    plt.figure(1); plt.subplot(122); plt.imshow(SLn_a[10:200,10:200],cmap=plt.cm.gray);plt.axis('off');plt.title('Average Filter -N = %d' %N);
    plt.show();
```

B.2 Repeat the operation with Gaussian filters 

Gaussian mask uses standard deviation
ranging from 1 to 10 with a step of 1;

Note the Gaussian mask size has to be at least 3x sigma at each direction

```python
for sigma in range(1,11):
    M = np.zeros((6*sigma+1,6*sigma+1));
    M[3*sigma,3*sigma]=1;
    M_g = filters.gaussian_filter(M,sigma);   # Gaussian Mask, note mask size have to be at least 3x sigma at each direction
    SLn_g = filters.convolve(SLn,M_g,mode='reflect');
    plt.figure(1); plt.subplot(121); plt.imshow(SLn_g,cmap=plt.cm.gray); plt.axis('off');plt.title('Gaussian Filter - STD = %d' %sigma);
    plt.figure(1); plt.subplot(122);plt.imshow(SLn_g[10:200,10:200],cmap=plt.cm.gray);plt.axis('off');plt.title('Gaussian Filter - STD = %d' %sigma);
    plt.show();
```

## C. Edge
C.1 Create the filters M_x and M_y corresponding to the two first order 
spatial derivatives

Apply them to SLn to obtain G_x and G_y, to compute the gradient magnitude
and the angle

Repeat for SLn_g (sigma = 1:10)

```python
M_x = filters.prewitt(np.array([[0,0,0],[0,1,0],[0,0,0]]));              # Mask gradient x direction
M_y = np.transpose(M_x);               #Mask gradient y direction
for sigma in range(0,11):
    if sigma>0:
        M = np.zeros((6*sigma+1,6*sigma+1));
        M[3*sigma,3*sigma]=1;
        M_g = filters.gaussian_filter(M,sigma);   # Gaussian Mask
        SLn_g = filters.convolve(SLn,M_g,mode = 'reflect');#Filtered SLN
    else:
        SLn_g = SLn;  # No Filter
    G_x = filters.convolve(SLn_g, M_x);          # Gradient in x direction
    G_y = filters.convolve(SLn_g, M_y);          # Gradient in y direction
    Mag = np.sqrt(G_x**2 + G_y**2);            # Gradient Magnitude
    Angle = np.arctan(G_y/G_x);         # Gradient Angle
    plt.figure(2);plt.subplot(221);plt.imshow(G_x,cmap=plt.cm.gray);plt.axis('off');plt.title('Gx');
    plt.subplot(222); plt.imshow(G_y,cmap=plt.cm.gray);plt.axis('off');plt.title('Gy'); 
    plt.subplot(223); plt.imshow(Mag,cmap=plt.cm.jet);plt.axis('off');plt.title('Magnitude');
    plt.subplot(224); plt.imshow(Angle,cmap=plt.cm.jet);plt.axis('off');plt.title('Angle');
    plt.show();
```

C.2 Create the Laplacian filter M_lap and apply it to SLn

Apply the LoG operator with standard deviation ranging from 0 
to 10 with a step of 1;

```python
M_lap = filters.laplace(np.array([[0,0,0],[0,1,0],[0,0,0]]));    # Laplacian Mask
Lap = filters.convolve(SLn, M_lap, mode = 'reflect');  # Filtered SLn
plt.figure(3);plt.subplot(211);plt.imshow(Lap,cmap=plt.cm.gray);plt.axis('off');plt.title('Lap');
for sigma in range(1,11):
    M = np.zeros((6*sigma+1,6*sigma+1));
    M[3*sigma,3*sigma]=1;
    M_log = filters.gaussian_laplace(M,sigma);    # LoG Mask
    Log = filters.convolve(SLn,M_log,mode = 'reflect'); # Filtered SLn
    plt.figure(3); plt.subplot(212);plt.imshow(Log,cmap=plt.cm.gray);plt.axis('off');plt.title('LoG Filter - STD = %d' %sigma);
    plt.show();
```

## D. Nonlinear filtering: edge-preserving filtering

D.1 Iterative median filter

Filter the image I_median with a median filter of size 5x5 (filter.median_filter) in an iterative fashion

```python
# Define number of time steps
steps = 50;
# plot noisy image
plt.figure(3);plt.subplot(221);plt.imshow(SLn,cmap=plt.cm.gray);plt.axis('off');plt.title('Noisy');
plt.subplot(222); plt.imshow(SLn[10:200,10:200],cmap=plt.cm.gray);plt.axis('off');plt.title('(region)');
# Make a copy for median filtered outcomes
I_median = SLn;
for i in range(1,steps):
    I_median = filters.median_filter(I_median,5);
    if np.mod(i,10)==1:
        plt.subplot(223); plt.imshow(I_median,cmap=plt.cm.gray);plt.axis('off');plt.title('Median filter');
        plt.subplot(224); plt.imshow(I_median[10:200,10:200],cmap=plt.cm.gray);plt.axis('off');plt.title('(region), iter = %d' %i);
        plt.show();   
```
