Import necessary libararies and do the necessary matplotlib magic:

```python
# the usual stuff
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt

# some adjustments and magic
np.set_printoptions(precision=4)
%matplotlib inline

# file system operations
import os

# for reading dicom files
import dicom
```

(1) Define a variable path which stores the name of the directory containg the dicom files, i.e. 'Aneurix_CAMP'. This data is a modified version of the freely available data set ANEURIX from Osirix (http://www.osirix-viewer.com/).

(2) Now use os in order to create a list of dicom files (dcm_file_list) in the aforementioned directory. HINT: Use tab completion to inspect the possibilities provided by os.

```python
# define path
path = 'Aneurix_CAMP'

# get list of all files
list_all_files = os.listdir(path)

# compose list of all dcm files
list_dcm_files = [file for file in list_all_files if file.endswith('.dcm')]
```

Determine number of slices (number of dcm files) using numpy and store it in a variable called slices.

```python
slices = np.size(list_dcm_files)
```

Read first dcm file in the list using dicom and determine its size (in pixels). Store the result in two variables called rows and cols.

```python
# read first dicom file for determining the dimension
Id = dicom.read_file(os.path.join(path,list_dcm_files[0]))

# get dimensions
rows = int(Id.Rows)
cols = int(Id.Columns)
```

(1) When working with 3D tomographic data it is very imporant to know that the intra-plane resolution is different from the inter-plane resolution (which is typically higher). Determine the physical resolution of the image data by finding and inspecting the respective DICOM tag.

(2) Browse other DICOM tags and find out the modality, the patient birth date, and the position of the image in 3D.

```python
# extract get pixel spacings in mm
pixelSpacing = (float(Id.PixelSpacing[0]), float(Id.PixelSpacing[1]), float(Id.SliceThickness))
pixelSpacing

# modality
#Id.Modality

# patient bith date
#Id.PatientBirthDate

# position
Id.ImagePositionPatient
```

(1) Create a numpy array called data of size rows x cols x slices filled with zeros.

(2) Loop over all files in the previously created list of dcm files, read them using dicom and store them in the created array.

```python
# create an array with the same datatype as the first image
data = np.zeros([rows, cols, slices], dtype=Id.pixel_array.dtype)

# loop through all the DICOM files
for file in list_dcm_files:
    
    # read the file
    Id = dicom.read_file(os.path.join(path,file))
    
    # store the raw image data
    data[:, :, list_dcm_files.index(file)] = Id.pixel_array
```

Problem: The order of the dcm filenames does not necessarily correspond to the position of the respective image! For this reason, it is important to determine the position of each slice in 3D before storing them in an array.

In order to sort your read DICOM data, get the image position of each DICOM image and sort it.

```python
# create array storing the image positions
imagePositions = np.zeros([slices])

# loop through all DICOM files
for file in list_dcm_files:
    
    # read the file
    Id = dicom.read_file(os.path.join(path,file))
    
    # store the raw image data
    imagePositions[list_dcm_files.index(file)] = Id.ImagePositionPatient[2];
    
# get list of sorted indices
indices = sp.argsort(imagePositions)

# re-sort array
data = data[:,:,indices[::-1]]
```

Display last slice again.

```python
fig = plt.plot(figsize=(20,10))
plt.imshow(data[:,:,slices-1], cmap=plt.cm.gray)
```

In order to explore the data even further, write yourself a small function called windowExplore which takes the volumetric data and an intensity range as input, converts the data to double, sets all pixel values outside the range to 0, and integrates all values in coronal direction up. 

```python
def windowExplore( volume, range ):

    # make a deep copy (and not a reference copy)
    vol = np.copy(volume)
    
    # find indices
    index_low = vol < range[0]
    index_high = vol > range[1]
    vol[index_low] = 0
    vol[index_high] = 0
    
    vol= np.sum(vol,0)
    return vol
```

Finally, we can use this function in oder to explore various aspects of the data. Create two subplots and visualize the result of windowExplore applied with the ranges [1250, 1350] and [1300, 4000]. Which tissue type becomes visible?

```python
# create figure and 1x2 array of axes
fig, axes = plt.subplots(1,2, figsize=(20,10))

# vasculature
vasc = windowExplore( data, [1250, 1350] )
axes[0].imshow(vasc, cmap=plt.cm.gray)
axes[0].set_title('vasculature')

# bones
bones = windowExplore( data, [1300, 4000] )
axes[1].imshow(bones, cmap=plt.cm.gray)
axes[1].set_title('bones')
```
