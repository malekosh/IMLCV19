# __Basic Image Segmentation__
 
## Thresholding

The goal of theses exercises is to become familiar with the basics image segmentation algorithms. At first, we import the usual stuff and some segmentation specific functions:

```python
import numpy as np
import os

import matplotlib
import matplotlib.pyplot as plt

from skimage.data import imread
from skimage.filters import threshold_otsu

np.set_printoptions(precision=4)
%matplotlib inline

matplotlib.rcParams['font.size'] = 12
```

Use the function imread to load the image 'data/TomograScanThorax.png', store it in the variable image and visualize this image.

```python
image = imread(os.path.join('data','TomograScanThorax.png'))

fig = plt.figure(figsize=(5, 5))
plt.title('Original')
plt.imshow(image, cmap=plt.cm.gray)
```

Get the histogram of the image and select a threshold value.

```python
plt.hist(image.ravel(), bins=256, range =(0.0, 255.0) )
plt.title('Histogram')
plt.axis('tight')

thresh = 158
plt.axvline(thresh, color='r')
```

Now print only a small region of the image.

```python
image[150:200,80:150] # portion of the image
```

Get a binary image with the threshold value selected and visualize the result

```python
binary = image > thresh
print(binary[150:200,80:150])

plt.imshow(binary ,cmap=plt.cm.gray)
```

Apply the Otsu's method and compare it with the previous results:

```python
thresh = threshold_otsu(image)
print('Otsu = ',thresh)
binary = image > thresh
plt.title('Binary image with Otsu')
plt.imshow(binary ,cmap=plt.cm.gray)
```

### k-means(k=2) vs Otsu
Load "data2.mat" and obtain the threshold value using Otsu's method. Apply k-means clustering (k = 2) with different initial centers. Is there any difference between the methods? What can you infer according to these results - and why?

```python
import scipy.io as sp

#Load data.mat
Data = sp.loadmat(os.path.join('data','data2.mat')) #Im = sp.loadmat('data/data2.mat')
#print(Data)
Im = Data['data']

thresh = threshold_otsu(Im)

fig = plt.figure(figsize=(16, 4))
ax1 = plt.subplot(1, 3, 1, adjustable='box-forced')
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3, sharex=ax1, sharey=ax1, adjustable='box-forced')

ax1.imshow(Im, cmap=plt.cm.gray)
ax1.set_title('Original')
ax1.axis('off')

ax2.hist(Im.ravel(), bins=256, range =(0.0, 255.0))
ax2.set_title('Histogram')
ax2.axvline(thresh, color='r')
ax2.axis('tight')

binaryOtsu = Im > thresh
ax3.imshow(binaryOtsu, cmap=plt.cm.gray)
ax3.set_title('Thresholded Otsu')
ax3.axis('off')
```

Additional imports for clustering

Documentation available:

- sklearn: http://scikit-learn.org/stable/modules/clustering.html#clustering

```python
# K-means

from sklearn.cluster import KMeans

c = np.shape(Im)
Im_reshape = np.reshape(Im, (c[0]*c[1],1))

centerManual = np.array([[120],[170]]) # seed
kmeans = KMeans(n_clusters = 2, n_init = 1, init = centerManual ).fit(Im_reshape)
print('Cluster centers: ',(kmeans.cluster_centers_))

Bw_kmeans = np.reshape(kmeans.labels_, (c[0],c[1]) )

fig = plt.figure(figsize=(16, 4))
ax1 = plt.subplot(1, 3, 1, adjustable='box-forced')
ax2 = plt.subplot(1, 3, 2, sharex=ax1, sharey=ax1, adjustable='box-forced')
ax3 = plt.subplot(1, 3, 3, sharex=ax1, sharey=ax1, adjustable='box-forced')

ax1.imshow(Im, cmap=plt.cm.gray)
ax1.set_title('Original')
ax1.axis('off')

ax2.imshow(binaryOtsu, cmap=plt.cm.gray)
ax2.set_title('Otsu')
ax2.axis('off')

ax3.imshow(Bw_kmeans, cmap=plt.cm.gray)
ax3.set_title('K_means')
ax3.axis('off')
```

### Multi-threshod

Apply k-means clustering (k > 2) with different initial centers to the image 'TomograScanThorax.png'

what can you infer according to these results?

```python

image = imread(os.path.join('data','TomograScanThorax.png'))

c = np.shape(image)
Im_reshape = np.reshape(image, (c[0]*c[1],1))

centerManual = np.array([[120],[170],[200]]) # seed
kmeans = KMeans(n_clusters = 3, n_init = 1, init = centerManual ).fit(Im_reshape)
print('Cluster centers: ',(kmeans.cluster_centers_))

Bw_kmeans = np.reshape(kmeans.labels_, (c[0],c[1]) )

fig = plt.figure(figsize=(16, 4))
ax1 = plt.subplot(1, 3, 1, adjustable='box-forced')
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3, sharex=ax1, sharey=ax1, adjustable='box-forced')

ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Original')
ax1.axis('off')

ax2.hist(image.ravel(), bins=256, range =(0.0, 255.0))
ax2.set_title('Histogram')
ax2.axvline(kmeans.cluster_centers_[0], color='r')
ax2.axvline(kmeans.cluster_centers_[1], color='b')
ax2.axvline(kmeans.cluster_centers_[2], color='g')
ax2.axis('tight')

ax3.imshow(Bw_kmeans, cmap=plt.cm.gray)
ax3.set_title('Thresholded Otsu')
ax3.axis('off')
```

### Region Growing

Now, let us implement the region growing method discussed in the lecture. The following function implements already a part of the method, i.e. the selection of potential neighborhood candidates.

```python
def get4ngb(rows, cols, x, y):
    # function ngb = get4ngb(rows, cols, x, y)
    # 
    # Input:
    # rows, cols    Number of rows and colums from the 2D Image
    # x,y           Coordinates of the center point
    #
    # Output:
    # ngb           list of possible neighbours

    # initialize empty result list
    ngb = list()

    # north
    if x > 1:
        ngb += [[x-1, y]];

    # east
    if y < cols-1:
        ngb += [[x, y+1]];


    # south
    if x < rows-1:
        ngb += [[x+1, y]];


    # west
    if y > 1:
        ngb += [[x, y-1]];

    return ngb 
```

Next, we have to define the main function for region growing. A rough skeleton is already given:

```python
def regionGrow(I, seedX, seedY, t):
    # function R = region_grow(I, t)
    #
    # I         Grayscale 2D Image
    # t         Threshold for deviation from seed point intensity
    # R         Resulting segmentation, a 2D matrix with the same size as I,
    #           (-1 = not yet enqueued, 0 = enqueued, 1 = segmented)
    #
    # Hint: Initialize this array at the beginning and use it during the algorithm

    # convert input image to double

    # store images rows and columns
    rows = np.size(I,0)
    cols = np.size(I,1)

    # initialize segmentation image labeling all pixels as not yet enqueued
    R = -np.ones((rows, cols))

    # initialize queue list with seed point
    queue = [[seedX, seedY]]
    

    # label seed point as enqueued in R
    q = queue[0]
    R[q[0], q[1]] = 0
    
    # loop for working on the queue
    while len(queue) > 0:
        
        # pick up and remove first pixel from queue
        q = queue[0]
        
        # check difference to seed point       
        if (np.abs(float(I[q[0], q[1]]) - float(I[seedX, seedY])) < t):

            # label pixel as segmented in R
            R[q[0], q[1]] = 1

            # determine neighbors
            nbgs = get4ngb(rows, cols, q[0], q[1]);

            # add neighbors to end of queue if not yet enqueued
            for i in range(0,len(nbgs)):
                
                qq = nbgs[i]
                if (R[qq[0], qq[1]] < 0):
                    
                    # enqueue it 
                    queue.append([qq[0], qq[1]])
                    
                    # label as enqueued
                    R[qq[0], qq[1]] = 0

        # remove first element
        queue = queue[1:]

    return R
```

Define a seed point, e.g. for segmenting one half of the lung and a maximum devitation from the seed point intensity and execute region growing:

```python
seed = [200, 400] # [coronal, sagital]
threshold = 50

R = regionGrow(image, seed[0], seed[1], threshold)

fig, axes = plt.subplots(1,2, figsize=(20,10))

# axial plane
axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('axial plane (input)')

# sagittal plane
axes[1].imshow(R, cmap=plt.cm.gray)
axes[1].set_title('segmentation')
```
